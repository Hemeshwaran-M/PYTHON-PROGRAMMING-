```python
#Bubble sort qstn 1 & 2
a = [75, 90, 65, 85, 70]

n = len(a)

for i in range(n):
    for j in range(n-i-1):
        if a[j] > a[j+1]:
            temp = a[j]
            a[j] = a[j+1]
            a[j+1] = temp

print("Sorted list:", a)
```

```python
#selection sort qstn 3 & 4 & 5
a = [88, 72, 95, 60, 81]

n = len(a)

for i in range(n):
    min_index = i

    for j in range(i+1, n):
        if a[j] < a[min_index]:
            min_index = j

    temp = a[i]
    a[i] = a[min_index]
    a[min_index] = temp

print("Sorted list:", a)
```

```python
#insertion sort qstn 6 & 7 & 17
a = [7, 3, 9, 1, 5]

n = len(a)

for i in range(1, n):
    key = a[i]
    j = i - 1

    while j >= 0 and key < a[j]:
        a[j+1] = a[j]
        j = j - 1

    a[j+1] = key 

print("Sorted list:", a)
```

```python
#Merge Sort qstn 9 & 10 & 18
a = [55, 12, 89, 34, 23, 78]

a1 = a[:3]
a2 = a[3:]

a1.sort()
a2.sort()

result = []

i = 0
j = 0

while i < len(a1) and j < len(a2):
    if a1[i] < a2[j]:
        result.append(a1[i])
        i += 1
    else:
        result.append(a2[j])
        j += 1

while i < len(a1):
    result.append(a1[i])
    i += 1

while j < len(a2):
    result.append(a2[j])
    j += 1

print("Sorted list:", result)
```

```python
#quick Sort qstn 8 & 11 & 12 & 13 & 14 & 15 & 16 & 19 & 20
a = [1000, 200, 1500, 500, 800]

pivot = a[0]

left = []
right = []

for i in range(1, len(a)):
    if a[i] <= pivot:
        left.append(a[i])
    else:
        right.append(a[i])

left.sort()
right.sort()

result = left + [pivot] + right

print("Sorted list:", result)
```
