```python
# 1. Half Pyramid Pattern
print("1. Half Pyramid")
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end="")
    print()

print("\n----------------")


# 2. Inverted Half Pyramid Pattern
print("2. Inverted Half Pyramid")
for i in range(5, 0, -1):
    for j in range(1, i + 1):
        print(j, end="")
    print()

print("\n----------------")


# 3. Repeated Number Pyramid
print("3. Repeated Number Pyramid")
for i in range(1, 6):
    for j in range(i):
        print(i, end="")
    print()

print("\n----------------")


# 4. Inverted Repeated Number Pyramid
print("4. Inverted Repeated Number Pyramid")
for i in range(5, 0, -1):
    for j in range(i):
        print(i, end="")
    print()

print("\n----------------")


# 5. Floyd's Triangle
print("5. Floyd's Triangle")
num = 1

for i in range(1, 6):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()

print("\n----------------")


# 6. Reverse Number Triangle
print("6. Reverse Number Triangle")
for i in range(1, 6):
    for j in range(i, 0, -1):
        print(j, end="")
    print()

print("\n----------------")


# 7. Right Shifted Number Triangle
print("7. Right Shifted Number Triangle")
for i in range(1, 6):
    for j in range(i, 6):
        print(j, end="")
    print()

print("\n----------------")


# 8. Palindrome Number Pyramid
print("8. Palindrome Number Pyramid")

for i in range(1, 6):

    # spaces
    for j in range(5 - i):
        print(" ", end="")

    # ascending
    for j in range(1, i + 1):
        print(j, end="")

    # descending
    for j in range(i - 1, 0, -1):
        print(j, end="")

    print()

print("\n----------------")


# 9. Continuous Number Triangle
print("9. Continuous Number Triangle")

num = 1

for i in range(1, 5):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()

print("\n----------------")


# 10. Binary Number Pattern
print("10. Binary Number Pattern")

for i in range(1, 6):
    for j in range(1, i + 1):

        if (i + j) % 2 == 0:
            print(1, end="")
        else:
            print(0, end="")

    print()
```
