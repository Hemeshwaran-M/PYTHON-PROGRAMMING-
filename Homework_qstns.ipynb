```python
# 88. Merge Sorted Array
nums1 = [1, 2, 3, 0, 0, 0]
m = 3
nums2 = [2, 5, 6]
n = 3

nums1[m:m+n] = nums2
nums1.sort()
print("88. Merge Sorted Array:", nums1)

# 169. Majority Element
nums = [2, 2, 1, 1, 1, 2, 2]
majority = max(set(nums), key=nums.count)
print("169. Majority Element:", majority)

# 217. Contains Duplicate
nums = [1, 2, 3, 1]
print("217. Contains Duplicate:", len(nums) != len(set(nums)))

# 242. Valid Anagram
s = "anagram"
t = "nagaram"
print("242. Valid Anagram:", sorted(s) == sorted(t))

# 268. Missing Number
nums = [3, 0, 1]
n = len(nums)
missing = n * (n + 1) // 2 - sum(nums)
print("268. Missing Number:", missing)
```

```python
# 1. Reverse a string without slicing
s = input("Enter a string for reverse: ")
rev = ""
for ch in s:
    rev = ch + rev
print("Reversed String:", rev)

# 2. Check palindrome without using reverse
s = input("\nEnter a string for palindrome check: ")
flag = True
for i in range(len(s) // 2):
    if s[i] != s[len(s) - i - 1]:
        flag = False
        break
if flag:
    print("Palindrome")
else:
    print("Not Palindrome")

# 3. Count vowels and consonants
s = input("\nEnter a string to count vowels and consonants: ").lower()
vowels = 0
consonants = 0
for ch in s:
    if ch.isalpha():
        if ch in "aeiou":
            vowels += 1
        else:
            consonants += 1
print("Vowels:", vowels)
print("Consonants:", consonants)

# 4. Find duplicate characters
s = input("\nEnter a string to find duplicate characters: ")
duplicates = []
for ch in s:
    if s.count(ch) > 1 and ch not in duplicates:
        duplicates.append(ch)
print("Duplicate Characters:", duplicates)

# 5. Remove special characters
s = input("\nEnter a string to remove special characters: ")
result = ""
for ch in s:
    if ch.isalnum() or ch.isspace():
        result += ch
print("After Removing Special Characters:", result)
```
