```python
# 1. Reverse a string without slicing
s = input("Enter a string for reverse: ")

rev = ""
for char in s:
    rev = char + rev

print("Reversed string:", rev)


# 2. Check palindrome without using reverse
s = input("\nEnter a string to check palindrome: ")

left = 0
right = len(s) - 1
is_palindrome = True

while left < right:
    if s[left] != s[right]:
        is_palindrome = False
        break
    left += 1
    right -= 1

if is_palindrome:
    print("Palindrome")
else:
    print("Not Palindrome")


# 3. Count vowels and consonants
s = input("\nEnter a string to count vowels and consonants: ").lower()

vowels = 0
consonants = 0

for char in s:
    if char.isalpha():
        if char in "aeiou":
            vowels += 1
        else:
            consonants += 1

print("Vowels:", vowels)
print("Consonants:", consonants)


# 4. Find duplicate characters
s = input("\nEnter a string to find duplicate characters: ")

count = {}

for char in s:
    if char in count:
        count[char] += 1
    else:
        count[char] = 1

print("Duplicate characters:")
for char in count:
    if count[char] > 1:
        print(char)


# 5. Remove special characters
s = input("\nEnter a string to remove special characters: ")

result = ""

for char in s:
    if char.isalnum() or char == " ":
        result += char

print("String after removing special characters:", result)
```
